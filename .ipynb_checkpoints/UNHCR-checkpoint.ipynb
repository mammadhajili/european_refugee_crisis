```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly
plotly.offline.init_notebook_mode()
```

```python
data = pd.read_json('UNHCR.json')
data.head()

```

```python
refugees = data[['country_code', 'country_iso', 'ref', 'year']]
refugees.head()
```

```python
grouped = refugees.groupby(['year']).sum()
grouped.head()
```

```python
plt.plot(grouped)
plt.show()
```

```python
eu = pd.read_csv("EU.csv")
eu.head()
```

```python
eu_refugees = pd.merge(refugees, eu, left_on='country_iso', right_on='ISO', how='inner')
eu_refugees.head()
```

```python
grouped = eu_refugees.groupby(['year']).sum()
plt.plot(grouped)
plt.show()
```

```python
data = []
years = eu_refugees[['year']].drop_duplicates().values

# let's create some additional, random data
for i in range(years.shape[0]):
    year_data = eu_refugees.loc[eu_refugees['year'] == years[i][0]]
    data.append(dict(type='choropleth',
             locations = year_data['ISO'].astype(str),
             z=year_data['ref'].astype(float),
             locationmode='ISO-3'))
    
# let's create the steps for the slider
steps = []
for i in range(len(data)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data)],
                label='Year {}'.format(i + 1951))
    step['args'][1][i] = True
    steps.append(step)

sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)]    
layout = dict(geo=dict(scope='europe', projection ={'type':'eckert4'}),
              sliders=sliders)

fig = dict(data=data, 
           layout=layout)
plotly.offline.iplot(fig)
```
